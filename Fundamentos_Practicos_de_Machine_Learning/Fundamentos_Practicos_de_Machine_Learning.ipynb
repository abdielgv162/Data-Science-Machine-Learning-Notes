### Sobreajuste o subajuste en los datos

Cuando entrenamos a nuestro algoritmo de aprendizaje lo que estamos haciendo es "oblig谩ndolo" a ajustarse a los datos de entrada y salida. Y hay que tener cuidado con 2 problemas, que suelen ser especialmente recurrentes al tratar con conjuntos de datos muy peque帽os.

Por un lado tenemos el **sobreajuste** o **overfitting** qu茅 es cuando entrenamos nuestro modelo de tal forma que **solo puede resolver un caso particular**, es cuando entrenamos el modelo de forma que tratamos de que se acople perfectamente a los datos de entrenamiento, cosa que puede parecer como algo que queremos lograr. El problema viene cuando le presentamos un nuevo set de datos y las predicciones que realiza sobre ellos ya no son tan acertadas.

Un ejemplo de esto es un algoritmo de detecci贸n de im谩genes, por ejemplo de gatitos, en donde en nuestro data set de entrenamiento solo le pasamos datos de gatos de color blanco de forma que el modelo se ajusta para detectar solo los patrones que componen a esos gatos blancos y al momento de pasarle un nuevo dato de por ejemplo, un gato negro, no lo va a detectar por que solo est谩 entrenado para un caso particular donde son blancos.

![](https://i.imgur.com/eJVwVJZ.png)

Por otro lado tenemos el **subajuste** o **underfitting** qu茅 es cuando el modelo falla al realizar la predicci贸n **por falta de muestras suficientes**. Es decir, no est谩 generalizando el conocimiento ya que nuestro conjunto de datos puede ser muy peque帽o.

Por lo regular la regla a seguir es la siguiente. Dado un conjunto de datos se suele destinar entre el 70 y 75% de ellos para la fase de entrenamiento y el restante se utiliza para evaluar (realizar predicciones y calificarlas), es decir, introducirlos al algoritmo sin su etiqueta correspondiente para que el algoritmo le asigne esa etiqueta.

![](https://i.imgur.com/bktxQDY.png)

### Regresi贸n lineal simple y regresi贸n lineal m煤ltiple

#### Regresi贸n lineal simple

Algoritmo de aprendizaje supervisado que nos indica la tendencia de un conjunto de **datos continuos**, modelando una relaci贸n entre una variable dependiente $Y$ y una variable explicativa llamada $X$. Como es un algoritmo de tipo supervisado tenemos que proporcionarle los datos previamente etiquetados.

Nuestra regresi贸n se puede expresar de la forma:

$$Y_i = b + X_i$$

que no es m谩s que la ecuaci贸n general de la recta, donde $b$ representa nuestro corte con el eje $y$.

![](https://i.imgur.com/xcmMAZ2.png)

Por lo regular, en problemas pr谩cticos nos interesa relacionar m谩s de una variable ya que mientras m谩s complejo se torna un problema, requerimos de m谩s variables para describirlo mejor. En ese caso recurrimos a utilizar una regresi贸n lineal **m煤ltiple**.

#### Regresi贸n lineal m煤ltiple

Si nuestra problema tiene m谩s de 2 variables se le considera lineal m煤ltiple. Nos va a permitir encontrar la tendencia entre las variables, pero esta vez teniendo en cuenta muchas m谩s variables que entran en juego.

En este caso pasamos a un hiper-plano y solo funciona con **datos cuantitativos**.

En el modelo de la regresi贸n lineal m煤ltiple esperamos que los sucesos tengan una forma funcional como:

$$y_j = b_o + b_1 x_{1j} + b_2 x_{2j} + \dots + b_k x_{kj} + u_j$$

donde $y$ es nuestra variable end贸gena, $x$ las variables ex贸genas, $u$ los residuos y $b$ los coeficientes estimados del efecto marginal entre cada $x$ e $y$.

Lo que hacemos es calcular promedios (promedios de una variable, de una correlaci贸n, de una tendencia, de una funci贸n, de un ratio; promedios de variables estandarizadas, estudentizadas o refuncionalizadas con cualquier otra funci贸n) y su desviaci贸n t铆pica (error est谩ndar) y una vez que son calculados se interpretan conjuntamente.

Ya tenemos el dataset y el modelo, as铆 que esperamos que al introducirle datos de nuevos pacientes podemos determinar si el paciente en cuesti贸n tiene diabetes o no, pero hay que tener cuidado ya que podemos llegar a obtener falsos positivos o falsos negativos.

Una forma de evaluar nuestro modelo es usando una matriz de confusi贸n, como la que se muestra en la siguiente imagen. 

![](https://i.imgur.com/XBSlhc6.png)

La matriz de confusi贸n nos va a ayudar a evaluar el desempe帽o de un algoritmo de clasificaci贸n a partir del conteo de aciertos y errores en cada una de las clases del algoritmo.

Como su nombre lo indica tenemos una matriz que nos ayuda a evaluar la predicci贸n mediante positivos y negativos como en la figura de arriba. Dentro de ella tenemos:

* **Verdaderos positivos (VP)**: Son aquellos que fueron clasificados correctamente como positivos en el modelo.
* **Verdaderos negativos (VN)**: Corresponden a la cantidad de negativos que fueron clasificados correctamente como negativos por el modelo.
* **Falsos negativos**: Es la cantidad de positivos que fueron clasificados incorrectamente como negativos.
* **Falsos positivos**: Indica la cantidad de negativos que fueron clasificados incorrectamente como positivos.

Para entenderlo mejor, consideremos un ejemplo.

Un m茅dico tiene 4 pacientes y a cada uno se le solicit贸 un examen de sangre y por error el laboratorio realiz贸 tambi茅n un estudio de embarazo, cuando los pacientes llegan el m茅dico les da los resultados.

A la primera paciente le da la noticia de que est谩 embarazada y ello ya lo sab铆a dado que tiene 3 meses de embarazo, es decir, un Verdadero Positivo.

El siguiente paciente llega y le dice que no est谩 embarazada y es una clasificaci贸n evidente dado que es hombre (Verdadero Negativo).

El tercer paciente llega y los resultados le indican que no est谩 embarazada, sin embargo, tiene 4 meses de embarazo, es decir, se ha clasificado como un Falso Negativo.

Y por 煤ltimo el cuarto paciente recibe en sus estudios que est谩 embarazado, sin embargo, es hombre, por lo cual es algo imposible y estamos visualizando un Falso Positivo.

$$\text{Exactitud} = \frac{VP + VN}{\text{Total}}$$

$$\text{Tasa de error} = \frac{FP + FN}{\text{Total}}$$

Vamos a tratar de crear una matriz de confusi贸n en c贸digo

## rboles de decisi贸n

### 驴Qu茅 son los 谩rboles de decisi贸n?

Es una forma gr谩fica y anal铆tica que presenta sucesos y sus posibles consecuencias. Permite que un individuo o una organizaci贸n comparen posibles acciones entre s铆 seg煤n su costo, probabilidades y beneficios. Se pueden usar para dirigir un intercambio de ideas informal o trazar un algoritmo que anticipe matem谩ticamente la mejor opci贸n.

Un 谩rbol de decisi贸n, por lo general, comienza con un 煤nico nodo y luego se ramifica en resultados posible. Cada uno de esos resultados crea nodos adicionales, que se ramifican en otras posibilidades. Esto le da una forma similar a la de un 谩rbol.

![](https://i.imgur.com/m0XQXw0.png)

Hay 3 tipos de nodos:
* **Nodos de probabilidad**: Usualmente representado con un c铆rculo, muestra las probabilidades de ciertos resultados.
* **Nodos de decisi贸n**: Representado con un cuadrado, muestra una decisi贸n que se tomar谩.
* **Nodos terminales**: Muestra el resultado definitivo de una ruta de decisi贸n.


![](https://d2slcw3kip6qmk.cloudfront.net/marketing/pages/chart/seo/decision-tree/discovery/sample-decision-tree.svg)

**Ventajas**

* Claridad de los datos
* Tolerantes al ruido y valores faltantes
* Las reglas extra铆das permiten hacer predicciones

**Desventajas**

* Suele ser deficiente el proceso de delimitar las rutas de decisiones cuando el 谩rbol es muy extenso
* Tiende a tener problemas de sobreajuste 
* Podemos llegar a tener ramas poco significativas

**Creaci贸n de un 谩rbol de decisi贸n**

1. Comenzamos con la decisi贸n principal: Dibujamos un recuadro para representar el punto inicial y le asignamos una l铆nea desde ese recuadro hac铆a cada posible soluci贸n o acci贸n, teni茅ndolas correctamente etiquetadas.


2. Agregamos nodos de decisi贸n y probabilidad para expandir el 谩rbol de modo que:
    * Si otra decisi贸n es necesaria, dibujamos otro recuadro extra.
    * Si el resultado es incierto, dibujamos un circula (representando un nodo de probabilidad)
    * Si el problema est谩 resuelto por ahora, lo dejamos en blanco.
    Desde cada nodo de decisi贸n dibujamos las soluciones posibles.
    
    
3. Continuar con la expansi贸n hasta que la l铆nea alcance un extremo, usualmente en los gr谩ficos indicamos esto con tri谩ngulos.

**驴C贸mo podemos dividir el 谩rbol?**

* Ganancia de informaci贸n
* Dividir en peque帽os 谩rboles

**Optimizaci贸n del modelo**

* Evitar el sobreajuste
* Selecci贸n correcta de atributos
* Evitar campos nulos
