### 쯈u칠 es la predicci칩n de datos?

Algoritmos que se definen como "clasificadores" que identifican a qu칠 conjunto de categor칤as pertenecen los datos.
 
Algo muy importante en la predicci칩n de nuestros datos consiste en que la informaci칩n que le vamos a proporcionar este sumamente cuidada, es decir, verificar si tenemos datos nulos, si nuestras columnas de datos est치n completas e incluso, si comprendemos el problema a resolver para saber que modelo vamos a utilizar y como vamos a interpretar los valores.

Recordemos que ya sea que tenemos un modelo de aprendizaje supervisado o no supervisado vamos a requerir de un conjunto de datos de entrenamiento con diferentes atributos, para ello nos podremos apoyar de bases de datos libres que contienen grandes data sets de informaci칩n.

---

### Sobreajuste o subajuste en los datos

Cuando entrenamos a nuestro algoritmo de aprendizaje lo que estamos haciendo es "oblig치ndolo" a ajustarse a los datos de entrada y salida. Y hay que tener cuidado con 2 problemas, que suelen ser especialmente recurrentes al tratar con conjuntos de datos muy peque침os.

Por un lado tenemos el **sobreajuste** o **overfitting** qu칠 es cuando entrenamos nuestro modelo de tal forma que **solo puede resolver un caso particular**, es cuando entrenamos el modelo de forma que tratamos de que se acople perfectamente a los datos de entrenamiento, cosa que puede parecer como algo que queremos lograr. El problema viene cuando le presentamos un nuevo set de datos y las predicciones que realiza sobre ellos ya no son tan acertadas.

Un ejemplo de esto es un algoritmo de detecci칩n de im치genes, por ejemplo de gatitos, en donde en nuestro data set de entrenamiento solo le pasamos datos de gatos de color blanco de forma que el modelo se ajusta para detectar solo los patrones que componen a esos gatos blancos y al momento de pasarle un nuevo dato de por ejemplo, un gato negro, no lo va a detectar por que solo est치 entrenado para un caso particular donde son blancos.

![](https://i.imgur.com/eJVwVJZ.png)

Por otro lado tenemos el **subajuste** o **underfitting** qu칠 es cuando el modelo falla al realizar la predicci칩n **por falta de muestras suficientes**. Es decir, no est치 generalizando el conocimiento ya que nuestro conjunto de datos puede ser muy peque침o.

Por lo regular la regla a seguir es la siguiente. Dado un conjunto de datos se suele destinar entre el 70 y 75% de ellos para la fase de entrenamiento y el restante se utiliza para evaluar (realizar predicciones y calificarlas), es decir, introducirlos al algoritmo sin su etiqueta correspondiente para que el algoritmo le asigne esa etiqueta.

![](https://i.imgur.com/bktxQDY.png)

### Regresi칩n lineal simple y regresi칩n lineal m칰ltiple

#### Regresi칩n lineal simple

Algoritmo de aprendizaje supervisado que nos indica la tendencia de un conjunto de **datos continuos**, modelando una relaci칩n entre una variable dependiente $Y$ y una variable explicativa llamada $X$. Como es un algoritmo de tipo supervisado tenemos que proporcionarle los datos previamente etiquetados.

Nuestra regresi칩n se puede expresar de la forma:

$$Y_i = b + X_i$$

que no es m치s que la ecuaci칩n general de la recta, donde $b$ representa nuestro corte con el eje $y$.

![](https://i.imgur.com/xcmMAZ2.png)

Por lo regular, en problemas pr치cticos nos interesa relacionar m치s de una variable ya que mientras m치s complejo se torna un problema, requerimos de m치s variables para describirlo mejor. En ese caso recurrimos a utilizar una regresi칩n lineal **m칰ltiple**.

#### Regresi칩n lineal m칰ltiple

Si nuestra problema tiene m치s de 2 variables se le considera lineal m칰ltiple. Nos va a permitir encontrar la tendencia entre las variables, pero esta vez teniendo en cuenta muchas m치s variables que entran en juego.

En este caso pasamos a un hiper-plano y solo funciona con **datos cuantitativos**.

En el modelo de la regresi칩n lineal m칰ltiple esperamos que los sucesos tengan una forma funcional como:

$$y_j = b_o + b_1 x_{1j} + b_2 x_{2j} + \dots + b_k x_{kj} + u_j$$

donde $y$ es nuestra variable end칩gena, $x$ las variables ex칩genas, $u$ los residuos y $b$ los coeficientes estimados del efecto marginal entre cada $x$ e $y$.

Lo que hacemos es calcular promedios (promedios de una variable, de una correlaci칩n, de una tendencia, de una funci칩n, de un ratio; promedios de variables estandarizadas, estudentizadas o refuncionalizadas con cualquier otra funci칩n) y su desviaci칩n t칤pica (error est치ndar) y una vez que son calculados se interpretan conjuntamente.

#### Evaluando el modelo

Ya tenemos el dataset y el modelo, as칤 que esperamos que al introducirle datos de nuevos pacientes podemos determinar si el paciente en cuesti칩n tiene diabetes o no, pero hay que tener cuidado ya que podemos llegar a obtener falsos positivos o falsos negativos.

Una forma de evaluar nuestro modelo es usando una matriz de confusi칩n, como la que se muestra en la siguiente imagen. 

![](https://i.imgur.com/XBSlhc6.png)

La matriz de confusi칩n nos va a ayudar a evaluar el desempe침o de un algoritmo de clasificaci칩n a partir del conteo de aciertos y errores en cada una de las clases del algoritmo.

Como su nombre lo indica tenemos una matriz que nos ayuda a evaluar la predicci칩n mediante positivos y negativos como en la figura de arriba. Dentro de ella tenemos:

* **Verdaderos positivos (VP)**: Son aquellos que fueron clasificados correctamente como positivos en el modelo.
* **Verdaderos negativos (VN)**: Corresponden a la cantidad de negativos que fueron clasificados correctamente como negativos por el modelo.
* **Falsos negativos**: Es la cantidad de positivos que fueron clasificados incorrectamente como negativos.
* **Falsos positivos**: Indica la cantidad de negativos que fueron clasificados incorrectamente como positivos.

Para entenderlo mejor, consideremos un ejemplo.

Un m칠dico tiene 4 pacientes y a cada uno se le solicit칩 un examen de sangre y por error el laboratorio realiz칩 tambi칠n un estudio de embarazo, cuando los pacientes llegan el m칠dico les da los resultados.

A la primera paciente le da la noticia de que est치 embarazada y ello ya lo sab칤a dado que tiene 3 meses de embarazo, es decir, un Verdadero Positivo.

El siguiente paciente llega y le dice que no est치 embarazada y es una clasificaci칩n evidente dado que es hombre (Verdadero Negativo).

El tercer paciente llega y los resultados le indican que no est치 embarazada, sin embargo, tiene 4 meses de embarazo, es decir, se ha clasificado como un Falso Negativo.

Y por 칰ltimo el cuarto paciente recibe en sus estudios que est치 embarazado, sin embargo, es hombre, por lo cual es algo imposible y estamos visualizando un Falso Positivo.

$$\text{Exactitud} = \frac{VP + VN}{\text{Total}}$$

$$\text{Tasa de error} = \frac{FP + FN}{\text{Total}}$$

Vamos a tratar de crear una matriz de confusi칩n en c칩digo

````python
# Generamos nuestra matriz de confusi칩n

# Recibe 2 par치etros, la informaci칩n que usamos para testear y la predicci칩n que arroj칩 el modelo
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
cnf_matrix
````

## 츼rboles de decisi칩n

### 쯈u칠 son los 치rboles de decisi칩n?

Es una forma gr치fica y anal칤tica que presenta sucesos y sus posibles consecuencias. Permite que un individuo o una organizaci칩n comparen posibles acciones entre s칤 seg칰n su costo, probabilidades y beneficios. Se pueden usar para dirigir un intercambio de ideas informal o trazar un algoritmo que anticipe matem치ticamente la mejor opci칩n.

Un 치rbol de decisi칩n, por lo general, comienza con un 칰nico nodo y luego se ramifica en resultados posible. Cada uno de esos resultados crea nodos adicionales, que se ramifican en otras posibilidades. Esto le da una forma similar a la de un 치rbol.

![](https://i.imgur.com/m0XQXw0.png)

Hay 3 tipos de nodos:
* **Nodos de probabilidad**: Usualmente representado con un c칤rculo, muestra las probabilidades de ciertos resultados.
* **Nodos de decisi칩n**: Representado con un cuadrado, muestra una decisi칩n que se tomar치.
* **Nodos terminales**: Muestra el resultado definitivo de una ruta de decisi칩n.


![](https://d2slcw3kip6qmk.cloudfront.net/marketing/pages/chart/seo/decision-tree/discovery/sample-decision-tree.svg)

**Ventajas**

* Claridad de los datos
* Tolerantes al ruido y valores faltantes
* Las reglas extra칤das permiten hacer predicciones

**Desventajas**

* Suele ser deficiente el proceso de delimitar las rutas de decisiones cuando el 치rbol es muy extenso
* Tiende a tener problemas de sobreajuste 
* Podemos llegar a tener ramas poco significativas

**Creaci칩n de un 치rbol de decisi칩n**

1. Comenzamos con la decisi칩n principal: Dibujamos un recuadro para representar el punto inicial y le asignamos una l칤nea desde ese recuadro hac칤a cada posible soluci칩n o acci칩n, teni칠ndolas correctamente etiquetadas.


2. Agregamos nodos de decisi칩n y probabilidad para expandir el 치rbol de modo que:
    * Si otra decisi칩n es necesaria, dibujamos otro recuadro extra.
    * Si el resultado es incierto, dibujamos un circula (representando un nodo de probabilidad)
    * Si el problema est치 resuelto por ahora, lo dejamos en blanco.
    Desde cada nodo de decisi칩n dibujamos las soluciones posibles.
    
    
3. Continuar con la expansi칩n hasta que la l칤nea alcance un extremo, usualmente en los gr치ficos indicamos esto con tri치ngulos.

**쮺칩mo podemos dividir el 치rbol?**

* Ganancia de informaci칩n
* Dividir en peque침os 치rboles

**Optimizaci칩n del modelo**

* Evitar el sobreajuste
* Selecci칩n correcta de atributos
* Evitar campos nulos

## K-Means

**k-means es un algoritmo de clasificaci칩n no supervisada** (clusterizaci칩n) que agrupa objetos en $k$ grupos bas치ndose en sus caracter칤sticas. El agrupamiento se realiza minimizando la suma de distancias entre cada objeto y el centroide de su grupo o cluster. Para este m칠todo solemos utilizar la distancia cuadr치tica.

El algoritmo consta de 3 pasos:

1. **Inicializaci칩n:** una vez que hemos elegido el n칰mero de grupos $k$, se establecen $k$ centroides en el espacio de los datos, por ejemplo, escogi칠ndolos aleatoriamente.

2. **Asignaci칩n de los objetos a los centroides:** cada objeto de los datos es asignado a su centroide m치s cercano.

3. **Actualizaci칩n de los centroides:** se actualiza la posici칩n del centroide de cada grupo tomando como nuevo centroide la posici칩n promedio de los objetos pertenecientes a dicho grupo.

Repetimos los pasos 2 y 3 hasta que los centroides no se mueven o se mueven por debajo de una distancia umbral.

Lo que hace el algoritmo *k-means* es resolver un problema de **optimizaci칩n**, siendo la funci칩n a optimizar, o minimizar en este caso; la suma de las distancias cuadraticas de cada objeto al centroide respectivo de su cluster.

Dado un conjunto de observaciones ($x_1, x_2, \dots, x_n )$, donde cada observaci칩n es un vector real de $d$ dimensiones, $k$-means construye una partici칩n de las observaciones en $k$ conjuntos ($k \le n$) a fin de minimizar la suma de los cuadrados dentro de cada grupo (minimizar la suma de la distancia de los objetos), dentro de cada grupo  $s= \{s_1,s_2, \cdots, s_k\}$, a su centroide. El problema se suele formular de la siguiente forma:

$$\underset{\mathbf{S}}{\mathrm{min}}\; E\left(\boldsymbol{\mu_{i}}\right)=\underset{\mathbf{S}}{\mathrm{min}}\sum_{i=1}^{k}\sum_{\mathbf{x}_{j}\in S_i}\left\Vert \mathbf{x}_{j}-\boldsymbol{\mu}_{i}\right\Vert ^{2}
\quad (1)$$

donde $S$ es el conjunto de datos cuyos elementos son los objetos $x_j$ representados por vectores, donde cada uno de sus elementos representa una caracter칤stica o atributo. Tendremos $k$ grupos o **clusters** con su correspondiente centroide $\mu_i$

### 쮺칩mo elegir el valor de k?

En general, no hay una forma exacta de determinar el n칰mero de grupos, pero se pueden usar ciertas reglas o estad칤sticos que nos ayudan a estimar el n칰mero de $k$.



#### 1. M칠todo del codo

La idea b치sica de los algoritmos de clustering es la **minimizaci칩n de la varianza intra-cluster** y la **maximizaci칩n de la varianza inter-cluster**. Es decir, queremos que cada observaci칩n se encuentre muy cerca a las de su mismo grupo y los grupos lo m치s lejor posible entre ellos.

El m칠todo del codo utiliza la distancia media de las observaciones a su centroide. Es decir, se fija en las distancias intra-cluster. Cuanto **m치s grande** es el n칰mero de clusters $k$, la **varianza intra-cluster tiende a disminuir**. Cuanto menor es la distancia intra-cluster mejor, ya que significa que los cl칰sters son m치s compactos. El m칠todo del codo busca el valor $k$ que satisfaga que un incremento de $k$ no mejore sustancialmente la distancia media intra-cluster.

![](https://i.imgur.com/CSaQUdo.png)

El m칠todo del codo es a veces ambiguo, una alternativa es el an치lisis de la silueta, que es m치s objetivo que el m칠todo del codo.

#### 2. An치lisis de la silueta

El an치lisis de la silueta mide la calidad del agrupamiento o clustering. Mide la distancia de separaci칩n entre los cl칰sters y nos indica como de cerca est치 cada punto de un cl칰ster a puntos de los cl칰sters vecinos. Esta medida de distancia se encuentra en el rango $[-1,1]$ donde un valor alto indica un buen clustering.

Los coeficientes de silueta cercanos a $+1$ indican que la observaci칩n **se encuentra lejos de los clusters vecinos.** Un valor del coeficiente de $0$ indica que la observaci칩n **est치 muy cerca o en las fronteras de decisi칩n entre 2 clusters**. Mientras que valores negativos indican que esas muestras quiz치s est칠n asignadas al **cl칰ster err칩neo**.

El m칠todo de la silueta calcula la media de los coeficientes de siluetas de todas las observaciones para diferentes valores de $k$. El n칰mero optimo de cl칰sters $k$ es aquel que maximiza la media de los coeficientes de silueta pra un rango de valores de $k$.

El coeficiente de la silueta es calculado como:

$$S= \frac{b-a}{\text{max}(a,b)}$$

siendo *a* la distancia media **intra-cluster** y *b* la **distancia media a las observaciones del cl칰ster m치s cercano**.

````python
wcss = [] # (within-cluster sums of squares)


for i in range (1, 11): # N칰mero de K's a evaluar
    kmeans = KMeans(n_clusters = i # Llamamos a evaluar al KMeans en el rango elegido
                    , max_iter = 1000 
                    , random_state = 0) 
    kmeans.fit(x) 
    # .intertia_ --> Suma de distancias cuadradas de muestras a su centro de conglomerado m치s cercano.
    # las agregamos a la lista wcss
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11), wcss) #Dibuja X entre 1 y 10. En Y los valores de las sumas
plt.title('Elbow Method')
plt.grid()
plt.xlabel('# of clusters')
plt.ylabel('WCSS')
````
