### Funciones

Recordando r谩pidamente qu茅 es una funci贸n:

* **Dominio:** Conjunto de los elementos que definen la funci贸n, es decir, los elementos que se van a asociar con otro conjunto, usualmente denotado con $x$.

* **Rango:** Es el conjunto de elementos que son el resultado de la asociaci贸n del dominio bajo la relaci贸n, usualmente denotado por $y$.

![](https://i.imgur.com/F24Np1f.png)

$$f(x)=y$$

### Composici贸n de funciones

Cuando hablamos de composici贸n de funciones estamos haciendo referencia al resultado de aplicar 2 o m谩s funciones sucesivas sobre un mismo elemento $x$.

Siendo $f$:

$$\begin{aligned}
f: X \longrightarrow  Y&  \\ \\
x \longmapsto y& = f(x) 
\end{aligned}$$

y $g$:

$$\begin{aligned}
g: Y \longrightarrow  Z&  \\ \\
y \longmapsto z& = g(x) 
\end{aligned}$$

Donde la imagen o **rango** de $f$ **est谩 contenido en el dominio de $g$**. Definimos la composici贸n de $f$ y $g$ como:

$$
\begin{aligned}
(g \circ f): X \longrightarrow  Z&  \\ \\
x \longmapsto z& = g(f(x)) 
\end{aligned}
$$

> Notemos que se nombran en el orden de aplicaci贸n de la variable, no en el orden sucesivo de la representaci贸n.

Y a todos los elementos de $X$ se le asocia un elemento de $Z$ seg煤n: $z=g(f(x))$:

$$\begin{aligned}
X \longrightarrow Y  \longrightarrow  &Z  \\ \\
x \longmapsto y  =f(x) \longmapsto &z = g(f(x)) 
\end{aligned}$$

Se puede representar de manera gr谩fica usando la categor铆a de conjuntos, mediante un diagrama conmutativo

![](https://upload.wikimedia.org/wikipedia/commons/thumb/e/ef/Commutative_diagram_for_morphism.svg/200px-Commutative_diagram_for_morphism.svg.png)

O de una manera m谩s intuitiva como la siguiente

![](https://www.universoformulas.com/imagenes/matematicas/analisis/composicion-funciones.jpg)

> **NOTA:** El s铆mbolo $\circ$ se llama **operador de anillo**.

```python
def f(x):
    return x**2

def g(x):
    return np.sin(x)

def graph(f1, f2, x):
    fig, ax = plt.subplots()
    plt.xlim(-4, 4)
    plt.ylim(-1, 3)
    ax.grid()
    ax.plot(x,f1, color='#F93800')
    ax.plot(x,f2, color='#011EFE')
    plt.legend(['f(x)','g(x)'])
    plt.title("Funciones originales")
    ax.axhline(y=0, color='#000000')
    ax.axvline(x=0, color='#000000')
    
def graph2(f1, f2, x):
    fig, ax = plt.subplots()
    plt.xlim(-4, 4)
    plt.ylim(-1, 3)
    ax.grid()
    ax.plot(x,f1, color='#10CB0D')
    ax.plot(x,f2, color='#3E00FF')
    plt.legend(['fog','gof'])
    plt.title("Composici贸n de funciones")
    ax.axhline(y=0, color='#000000')
    ax.axvline(x=0, color='#000000')
    
    
if __name__ == '__main__':
    res = 10000
    x = np.linspace(-50.0, 50.0, num=res)
    y = f(x)
    z = g(x)
    f_o_g = f(g(x))
    g_o_f = g(f(x))
    display_graph = graph(y, z, x)
    display_graph = graph2(f_o_g, g_o_f, x)
```

**2. La funci贸n es estrictamente decreciente:**

**3. La funci贸n es mon贸tona y creciente:**

$$\text{si para todo } x_1, x_2 \in I: x_1 < x_2 \Longleftrightarrow f(x_1) \leq f(x_2)$$

**4. La funci贸n es mon贸tona y decreciente:**

$$\text{si para todo } x_1, x_2 \in I: x_1 < x_2 \Longleftrightarrow f(x_1) \geq f(x_2)$$

$$s(u)=

### Derivando

#### Suma

$$(f+g)\prime (x) = f\prime (x) + g\prime (x)$$

#### Producto

$$(f \cdot g)\prime(x) = f\prime(x) \cdot(g) + f(x)\cdot g\prime(x)$$

#### Cociente

$$\bigg(\frac{f(x)}{g(x)}\bigg)\prime = \frac{f\prime (x)g(x) - f(x)g\prime (x)}{(g(x))^2}$$

#### Regla de la cadena

Esta nos dice como podemos calcular la derivada de una composici贸n de funciones.

$$\begin{aligned}
(f \circ g)\prime (x) &= (f\prime \circ g)\cdot g\prime \\
&= f\prime(g(x)) \cdot g(x)\prime
\end{aligned}$$

> La derivada de una composici贸n de funciones es igual a la derivada de $f(x)$ evaluada en $g(x)$ ,multiplicada por la derivada de $g(x)$.

o en notaci贸n de Leibniz:

$$\frac{dz}{dx} = \frac{dz}{dy} \cdot \frac{dy}{dx}$$

> Si una variable $z$ depende de una variable $y$ y a su vez esta depende de una variable $x$ entonces, $z$ tambi茅n depende de $x$.

### Criterios de la 1ra y 2da derivada

#### Criterio de la primera derivada

Se le llama criterio de la primera derivada al m茅todo o teorema utilizado para determinar los m铆nimos y m谩ximos **relativos** que pueden existir, donde se observa el cambio de signo en un intervalo abierto se帽alado que contiene al punto cr铆tico $c$.

* Si $f$ es derivable en el intervalo, excepto posiblemente en $c$, entonces $f(c)$ puede clasificarse como sigue:
     * Si $f\prime(x)$ cambia de negativa a positiva en $c$, entonces $f$ tiene un m铆nimo relativo en $(c, f(c))$.
     
     * Si $f\prime(x)$ cambia de positiva a negativa en $c$, entonces $f$ tiene un m谩ximo relativo en $(c, f(c))$.
     
     * Si $f\prime(x)$ es positiva en ambos lados de $c$ o negativa en ambos lados de $c$, entonces $f(c)$ NO es ni un m铆nimo ni un m谩ximo.

##### Ejemplo

obtenemos su derivada

$$f\prime(x) = 3x^2 -2x - 1 $$

y para obtener los m谩ximos y m铆nimos locales igualamos a 0 y resolvemos

$$\begin{aligned}
3x^2 -2x - 1 &= 0  \\ \\ 
x^2 - \frac{2}{3} x +  \bigg(\frac{2}{6}\bigg)^2  &=  \frac{1}{3} + \bigg(\frac{2}{6}\bigg)^2 \\ \\
              x^2 - \frac{2}{3} x +  \frac{1}{9}  &=   \frac{4}{9} \\ \\
           \bigg(x -  \sqrt{\frac{1}{9}}\bigg)^2  &=   \frac{4}{9} \\ \\
           \sqrt{\bigg(x -  \frac{1}{3}\bigg)^2}  &=   \sqrt{\frac{4}{9}} \\ \\
                                 x -  \frac{1}{3} &=  \pm \bigg(\frac{2}{3}\bigg) \\ \\
                                              x_1 = -\frac{1}{3}&, \quad x_2= 1
\end{aligned}$$

finalmente, evaluamos los puntos cr铆ticos en nuestra funci贸n original

$$ f\bigg(-\frac{1}{3}\bigg) = x^3 - x^2 - x + 2 = \frac{59}{27}$$

$$ f(1) = x^3 - x^2 - x + 2 = 1$$

```python
import matplotlib.pyplot as plt
import numpy as np

def fx(x):
    return x**3 - x**2 - x + 2 

def f_prime(x):
    return (3* x**2) - (2* x) - 1

def graph(eq, eq2):
    fig, ax = plt.subplots()
    ax.plot(x,eq, color='#54D562')
    ax.plot(x,eq2, color='#788AFD')
    plt.xlim(-3, 3)
    plt.ylim(-2, 4)
    
    plt.plot(-1/3, 59/27, 'o', color ='#000000')
    plt.plot(1, 1, 'o', color ='#000000')
    ax.axhline(y=59/27, color='#0334C2', linestyle = 'dotted')  
    ax.axhline(y=1, color='#0334C2', linestyle = 'dotted')
    
    
    plt.legend(['f (x)','f \'(x) '])
    ax.grid()
    ax.axvline(x=0, color='#000000')  
    ax.axhline(y=0, color='#000000') 
    plt.title(f'M谩ximos y m铆nimos relativos con 1ra derivada')
              
if __name__ == '__main__':              
    res = 10000
    x = np.linspace(-10.0, 10.0, num=res)
    y = fx(x)
    y2 = f_prime(x)
    graph(y,y2)
```

#### Criterio de la segunda derivada

Se utiliza la segunda derivada para efectuar una prueba simple correspondiente a los m谩ximos y m铆nimos relativos. Se basa en el hecho de que si la gr谩fica de una funci贸n $f$ es c贸ncava hacia arriba en un intervalo abierto que contiene a $c$, entonces $f\prime(x)=0$ debe ser un m铆nimo relativo de $f$. De manera similar, si la gr谩fica de una funci贸n es c贸ncava hacia abajo en un intervalo abierto que contiene a $c$, entonces $f\prime(x)=0$ debe de ser un m谩ximo relativo de $f$. 

* Si $f\prime \prime(c)>0$, entonces $f$ tiene un m铆nimo relativo en $(c, f(c))$.
* Si $f\prime \prime(c)<0$, entonces $f$ tiene un m谩ximo relativo en $(c, f(c))$.
* Si $f\prime \prime(c)=0$, entonces el criterio falla. Esto es, $f$ quiz谩s tenga un m谩ximo relativo en $c$, un m铆nimo relativo en $(c,f\prime(c))$ o ninguno de los dos. En tales casos, se puede utilizar el criterio de la primera derivada o el criterio de la tercera derivada.

##### Ejemplo

Teniendo la funci贸n

$$f(x) = 3x^3 - 8x^2 +7x -2$$

con puntos cr铆ticos en

$$x=1 \text{ y } x=\frac{7}{9}$$

con primera derivada

$$f\prime(x)= 9x^2 - 16x + 7$$

y segunda derivada

$$f\prime \prime (x)= 18x - 16$$

evaluamos a $f\prime \prime$ en $f \prime \prime(1)= 18(1)-16 = 2$, obteniendo que $f\prime \prime (1) > 0$ y por lo tanto, **tenemos un m铆nimo local en $x=1$**

 y tambi茅n  $f\prime \prime (7/9)= 18(7/9)-16 = -2 $, obteniendo que $f \prime \prime (7/9) < 0$ y por lo tanto, **tenemos un m谩ximo local en $x=\frac{7}{9}$**.

```python
import matplotlib.pyplot as plt
import numpy as np

def fx(x):
    return 3*(x**3) - 8*(x**2) + 7*x - 2 

def f_prime(x):
    return 9*(x**2) - 16*x + 7

def f_prime_prime(x):
    return 18*x - 16

def graph(eq, eq2, eq3):
    fig, ax = plt.subplots()
    ax.plot(x,eq, color='#F978FD')
    ax.plot(x,eq2, color='#682CD9')
    ax.plot(x,eq3, color='#093CBA')
    plt.xlim(-0.25, 2)
    plt.ylim(-2, 3)
    # Puntos cr铆ticos evaluados en la 2da derivada
    plt.plot(1, 2, 'o', color ='#000000')
    plt.plot(7/9, -2, 'o', color ='#000000')
    # Puntos cr铆ticos evaluados en la 1ra derivada
    plt.plot(1, 0, 'o', color ='#000000')
    plt.plot(7/9, 0, 'o', color ='#000000')
    ax.axvline(x=1, color='#0334C2', linestyle = 'dotted')  
    ax.axvline(x=7/9, color='#0334C2', linestyle = 'dotted')
    plt.legend(['f (x)','f \'(x)', 'f \'\' (x)' ])
    ax.grid()
    ax.axvline(x=0, color='#000000')  
    ax.axhline(y=0, color='#000000') 
    plt.title(f'M谩ximos y m铆nimos relativos con 2da derivada')

    
if __name__ == '__main__':              
    res = 10000
    x = np.linspace(-10.0, 10.0, num=res)
    y = fx(x)
    y2 = f_prime(x)
    y3 = f_prime_prime(x)
    graph(y,y2, y3)
```
