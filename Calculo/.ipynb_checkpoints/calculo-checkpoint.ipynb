### Funciones

Recordando r谩pidamente qu茅 es una funci贸n:

Es una relaci贸n donde a cada elemento de un conjunto $A$ se le asigna un elemento de un conjunto $B$ siempre que se vinculen de manera 煤nica y exclusiva. En otras palabras, a cada valor del dominio, le corresponde **uno y s贸lo un valor del rango**.

* **Dominio:** Conjunto de los elementos que definen la funci贸n, es decir, los elementos que se van a asociar con otro conjunto, usualmente denotado con $x$.

* **Rango:** Es el conjunto de elementos que son el resultado de la asociaci贸n del dominio bajo la relaci贸n, usualmente denotado por $y$.

![](https://i.imgur.com/F24Np1f.png)

En una funci贸n, entra un elemento $x$ y sale un elemento $y$:

$$f(x)=y$$

#### Funciones a trozos

Son funciones que tienen diferentes valores definidos por un intervalo. Por ejemplo la funci贸n escal贸n de Heaviside:

$$H(x) = 
     \begin{cases}
        0, &\quad \text{para, } x < 0 \\
        1,  &\quad\text{para. } x \ge 0 \\
     \end{cases}
$$

```python
def f(x):
    # Obtenemos un vector lleno de 0's
    y = np.zeros(len(x))
    # Obtenemos el 铆ndice y los valores para rellenar con 1's
    for idx, x in enumerate(x):
        if x>= 0:
            y[idx] = 1
    return y


def graph(equation):
    fig, ax = plt.subplots()
    ax.plot(x,y, color='#D30FB8')
    plt.xlim(-3, 3)
    plt.ylim(-1, 3)
    plt.legend(['H(x)'])
    ax.grid()
    ax.axvline(x=0, color='#000000') 
    plt.plot(0, marker = 'o', ms = 15, mfc ='w')
    plt.plot(1, marker = 'o', ms = 15, mfc ='#FF00FF')
    plt.title(f'Gr谩fica de una funci贸n a trozos')
    
if __name__ == '__main__':
    res = 10000
    # Como es una variable usamos un vector 
    # Establecemos el intervalo y cuandos elementos queremos (res)
    x = np.linspace(-10.0, 10.0, num=res)
    y = f(x)
    display_graph = graph(y)
```

#### Desplazamientos verticales y horizontales

Siendo $c$ una constante mayor a 0, entonces la gr谩fica:

* $y=f(x) + c$ se desplaza $c$ unidades hacia arriba.
* $y=f(x) - c$ se desplaza $c$ unidades hacia abajo.
* $y=f(x - c) $ se desplaza $c$ unidades hacia la derecha.
* $y=f(x+ c) $ se desplaza $c$ unidades hacia la izquierda.

```python
def fs(x,a):
    # Establecemos la relaci贸n algebraica
    return x**2, (x**2)+a, (x**2)-a, (x-a)**2,(x+a)**2


def graph(f1, f2, f3, f4, f5, x):
    fig, ax = plt.subplots()
    plt.xlim(-15, 15)
    plt.ylim(-10, 100)
    ax.plot(x,f1, color='#D43434')
    ax.plot(x,f2, color='#FFB900')
    ax.plot(x,f3, color='#0821F3')
    ax.plot(x,f4, color='#7615DC')
    ax.plot(x,f5, color='#FF03E0')
    plt.legend(['x^2','x^2+a: c unidades hacia arriba ','x^2-a: c unidades hacia abajo.',
                '(x-a)^2: c unidades hacia la derecha', '(x+a)^2: c unidades hacia la izquierda'],
               bbox_to_anchor=(1, 1), loc='upper left')
    ax.grid()
    ax.axhline(y=0, color='#000000')
    ax.axvline(x=0, color='#000000')  
    plt.title("Desplazamientos verticales y horizontales")
    
if __name__ == '__main__':
    res = 10000
    a = int(input('Ingresa el valor de la constante a: '))
    x = np.linspace(-50.0, 50.0, num=res)
    f1, f2, f3, f4, f5 = fs(x,a)
    display_graph = graph(f1, f2, f3, f4, f5, x)
```

#### Alargamientos y compresiones

* $y=c\cdot f(x)$ alarga la gr谩fica verticalmente por un factor de $c$ unidades.
* $y=\frac{1}{c} \cdot f(x)$ comprime la gr谩fica verticalmente por un factor de $c$ unidades.
* $y=f(c \cdot x)$ comprime la gr谩fica horizontalmente por un factor de $c$ unidades.
* $y= f(\frac{1}{c}\cdot x )$ alarga la gr谩fica horizontalmente por un factor de $c$ unidades.

### Composici贸n de funciones

Cuando hablamos de composici贸n de funciones estamos haciendo referencia al resultado de aplicar 2 o m谩s funciones sucesivas sobre un mismo elemento $x$.

Siendo $f$:

$$\begin{aligned}
f: X \longrightarrow  Y&  \\ \\
x \longmapsto y& = f(x) 
\end{aligned}$$

y $g$:

$$\begin{aligned}
g: Y \longrightarrow  Z&  \\ \\
y \longmapsto z& = g(x) 
\end{aligned}$$

Donde la imagen o **rango** de $f$ **est谩 contenido en el dominio de $g$**. Definimos la composici贸n de $f$ y $g$ como:

$$
\begin{aligned}
(g \circ f): X \longrightarrow  Z&  \\ \\
x \longmapsto z& = g(f(x)) 
\end{aligned}
$$

> Notemos que se nombran en el orden de aplicaci贸n de la variable, no en el orden sucesivo de la representaci贸n.

Y a todos los elementos de $X$ se le asocia un elemento de $Z$ seg煤n: $z=g(f(x))$:

$$\begin{aligned}
X \longrightarrow Y  \longrightarrow  &Z  \\ \\
x \longmapsto y  =f(x) \longmapsto &z = g(f(x)) 
\end{aligned}$$

Se puede representar de manera gr谩fica usando la categor铆a de conjuntos, mediante un diagrama conmutativo

![](https://upload.wikimedia.org/wikipedia/commons/thumb/e/ef/Commutative_diagram_for_morphism.svg/200px-Commutative_diagram_for_morphism.svg.png)

O de una manera m谩s intuitiva como la siguiente

![](https://www.universoformulas.com/imagenes/matematicas/analisis/composicion-funciones.jpg)

> **NOTA:** El s铆mbolo $\circ$ se llama **operador de anillo**.

```python
def f(x):
    return x**2

def g(x):
    return np.sin(x)

def graph(f1, f2, x):
    fig, ax = plt.subplots()
    plt.xlim(-4, 4)
    plt.ylim(-1, 3)
    ax.grid()
    ax.plot(x,f1, color='#F93800')
    ax.plot(x,f2, color='#011EFE')
    plt.legend(['f(x)','g(x)'])
    plt.title("Funciones originales")
    ax.axhline(y=0, color='#000000')
    ax.axvline(x=0, color='#000000')
    
def graph2(f1, f2, x):
    fig, ax = plt.subplots()
    plt.xlim(-4, 4)
    plt.ylim(-1, 3)
    ax.grid()
    ax.plot(x,f1, color='#10CB0D')
    ax.plot(x,f2, color='#3E00FF')
    plt.legend(['fog','gof'])
    plt.title("Composici贸n de funciones")
    ax.axhline(y=0, color='#000000')
    ax.axvline(x=0, color='#000000')
    
    
if __name__ == '__main__':
    res = 10000
    x = np.linspace(-50.0, 50.0, num=res)
    y = f(x)
    z = g(x)
    f_o_g = f(g(x))
    g_o_f = g(f(x))
    display_graph = graph(y, z, x)
    display_graph = graph2(f_o_g, g_o_f, x)
```

#### Funciones mon贸tonas

Estas funciones son 煤tiles de reconocer o analizar debido a que **nos permite saber si una funci贸n crece o decrece en alguno de sus intervalos**. Que algo sea mon贸tono se refiere a que no tiene variaciones. Entonces, las funciones mon贸tonas son aquellas dentro de un intervalo $I \in \mathbb{R}$, que **cumplen con alguna** de las siguientes propiedades:

**1. La funci贸n es mon贸tona y estrictamente creciente:**

$$\text{si para todo } x_1, x_2 \in I: x_1 < x_2 \Longleftrightarrow f(x_1) < f(x_2)$$

> Si para todo $x_1$ y $x_2$ que pertenecen al intervalo $I$, tal que $x_1$ sea menor a $x_2$, si y solo si $f(x_1)$ es menor que $f(x_2)$.

Lo que nos est谩 diciendo esta definici贸n es que $x_1$ siempre tiene que ser menor que $x_2$ en nuestro intervalo $I$, y que al evaluar $x_2$ en la funci贸n obtendremos un n煤mero mayor que al evaluar $x_1$ (es creciente).

La siguientes definiciones se leen de una forma muy similar.

**2. La funci贸n es estrictamente decreciente:**

$$\text{si para todo } x_1, x_2 \in I: x_1 < x_2 \Longleftrightarrow f(x_1) > f(x_2)$$

**3. La funci贸n es mon贸tona y creciente:**

$$\text{si para todo } x_1, x_2 \in I: x_1 < x_2 \Longleftrightarrow f(x_1) \leq f(x_2)$$

**4. La funci贸n es mon贸tona y decreciente:**

$$\text{si para todo } x_1, x_2 \in I: x_1 < x_2 \Longleftrightarrow f(x_1) \geq f(x_2)$$

### 驴C贸mo se compone una neurona?

Una neurona artificial o neurona de McCulloch-Pitts , es una unidad de c谩lculo que intenta modelar el comportamiento de una neurona "natural", similares a las que constituyen al cerebro humano. En su nivel m谩s fundamental no es m谩s que una funci贸n matem谩tica, de manera que recibe inputs o **est铆mulos** para poder realizar su operaci贸n o **activarse**.

El nuestras entradas tendremos los valores $x_1, x_2, ..., x_n$. Cada una asociada con un peso $w_1, w_2, ..., w_n$. Adem谩s de que tendremos un par谩metro $b$ que representa nuestro threshold, que es un rango que le da cierta flexibilidad a nuestra neurona para saber c贸mo ajusta el valor.

$$y = f(x)= w_1x_1+ w_2x_2+ ... w_nx_n + b $$

tambi茅n podemos encontrar a esta suma ponderada denotada por $red$.

Gr谩ficamente lo que tenemos es una linea recta, donde nuestra combinaci贸n lineal $ w_1x_1+ w_2x_2+ ... w_nx_n $ toma el lugar de nuestra parte $mx$ de la ecuaci贸n general de la recta, mientras que el t茅rmino $b$ (bias) representa la intersecci贸n con el eje $y$.

### Funciones de activaci贸n en una neurona $\phi$

Ya tenemos una funci贸n dentro de cada neurona que corresponder谩 con la suma ponderada de sus entradas junto con el bias que nos permite un poco de flexibilidad para la funci贸n, pero a煤n tenemos que encontrar la forma de saber en que punto la neurona est谩 activada o no, estableciendo un rango de valores de umbral.

Usualmente de describen de la forma:

$$s(u)=

\end{cases}$$

> Se activa si $u \geq 0 $. Si nos damos cuenta, esta es la funci贸n de Heaviside que graficamos arriba.

Entonces al conectar nuestras neuronas obtendremos una composici贸n de funciones de la forma:

#### Tangente hiperb贸lica

Esta es otra funci贸n que se suelte utilizar en las funciones de activaci贸n. Esta representa un n煤mero real $x$ designado mediante $\tanh x$ y se define como el cociente entre el seno hiperb贸lico y el coseno hiperb贸lico del n煤mero real $x$.
